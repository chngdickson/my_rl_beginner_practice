The whole point of gradient Monte Carlo is rather than using Vpi(s) true value that is based on dynamic programming. We use an approximator to estimate our value.


We can see here that delta v = alpha*(Vpi(S)-V(S,w)
It deficits the error term.One may wonder why won't we make a bigger step like the original function that uses **Square** , why did we make delta v a square root as what it's supposed to be? The reason behind it is rather than obtaining zero error for all states, we would prefer the function approximation to balance the errors in different state. In layman terms "**Generalize**" rather than overfit all states. Hence create an unbiased estimate.

```python
N_STATES = 1000
class ValueFunction:
    # @num_of_groups: # of aggregations
    def __init__(self, num_of_groups):
        self.num_of_groups = num_of_groups
        self.group_size = N_STATES // num_of_groups

        # thetas
        self.params = np.zeros(num_of_groups)

    # get the value of @state
    def value(self, state):
        if state in END_STATES:
            return 0
        group_index = (state - 1) // self.group_size
        return self.params[group_index]

    # update parameters
    # @delta: step size * (target - old estimation)
    # @state: state of current sample
    def update(self, delta, state):
        group_index = (state - 1) // self.group_size
        self.params[group_index] += delta

# Each instance of aggregate has 100 states out of 1000 total states.        
value_function = ValueFunction(10)
```
